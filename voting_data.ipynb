```python
df=pd.read_csv('UNVotes.csv')
```

```python
df=df[['rcid', 'vote', 'Country', 'year']]
df=df[df['vote']!=8] # absent
df=df[df['vote']!=9] # not a member
df['vote'] = df['vote'].map({1: 1, 3: 0, 2: -1}) # to have {'yes': 1, 'no': 0, 'abstain': -1}
df=df.drop_duplicates(subset=['rcid', 'year','Country'])
df.to_csv('clean_UNVotes.csv')
df
```

```python
df_wide = pd.pivot_table(df, values='vote', index=['Country', 'year'], columns='rcid', aggfunc='first').reset_index()
df_wide_filled = df_wide.fillna(9)
df_wide_filled.to_csv('clean_UNVotes_pivoted.csv')
```
